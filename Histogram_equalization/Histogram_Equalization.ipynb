```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/dark image.jpeg', 0)

# Calculate the histogram of the original image
hist1 = cv2.calcHist([img], [0], None, [256], [0, 256])

# Create a new histogram after equalization
equalizedImage1 = cv2.equalizeHist(img)
equalizedHist1 = cv2.calcHist([equalizedImage1], [0], None, [256], [0, 256])

# Plotting
plt.subplot(221), plt.plot(hist1)
plt.subplot(222), plt.plot(equalizedHist1)
plt.subplot(223), plt.imshow(img, cmap='gray')
plt.subplot(224), plt.imshow(equalizedImage1, cmap='gray')

plt.show()

# Showing the result image
result1 = np.hstack((img, equalizedImage1))
# cv2_imshow( result1)

cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/dark image.jpeg', 0)
imgB = cv2.imread('/content/drive/MyDrive/Colab Notebooks/view.jpg', 0)

hist1 = cv2.calcHist([img], [0], None, [256], [0,256])
hist2 = cv2.calcHist([imgB], [0], None, [256], [0,256])

# Create new histogram after equalizer

equalizedImage1 = cv2.equalizeHist(img);
equalizedImageB = cv2.equalizeHist(imgB);

equalizedHist1 = cv2.calcHist([equalizedImage1], [0], None, [256], [0, 256])
equalizedHist2 = cv2.calcHist([equalizedImageB], [0], None, [256], [0, 256])

#plotting
plt.subplot(221), plt.plot(hist1)
plt.subplot(222), plt.plot(hist2)
plt.subplot(223), plt.plot(equalizedHist1)
plt.subplot(224), plt.plot(equalizedHist2)

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/yash.jpg", 1)

# convert image from rgb to hsv
img_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# histogram equalization on the v-channel
img_hsv[:,:,2] = cv2.equalizeHist(img_hsv[:,:,2])

# convert the image from hsv to rgb
imgRgb = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

# histogram
hist1 = cv2.calcHist([image], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([imgRgb], [0], None, [256], [0, 256])

#plotting
plt.subplot(221), plt.plot(hist1)
plt.subplot(222), plt.plot(hist2)
plt.subplot(223), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(224), plt.imshow(cv2.cvtColor(imgRgb, cv2.COLOR_BGR2RGB))

plt.show()
```
